```python
data_match = data_match.sort_values(by = ['Team','Date']).reset_index()
data_match
```

```python
data_match = data_match.sort_values(by='Goal Scored', ascending=False).reset_index()
data_match
```

```python
def get_stats(group):
    return {'min': group.min(), 'max': group.max(), 'mean': group.mean()}
```

```python
data_match['Goal Scored'].groupby(data_match['Team']).apply(get_stats).unstack()
```

```python
new_feature = data_match.merge(data_cou, left_on='Team', right_on='Country')
new_feature
```

```python
continent = new_feature.groupby(['Continent']).sum()
continent
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
viz_1 = pd.DataFrame(data_match.groupby(['Team','On-Target']) ['Goal Scored'].sum())
viz_1
```

```python
viz_1 = viz_1.reset_index()
viz_1.head()
```

```python
viz_1 = viz_1.groupby('Team').sum().reset_index().sort_values(by=['Goal Scored'], ascending = False)
viz_1.head()
```

```python
viz_1 = viz_1.set_index('Team')
viz_1.head()
```

```python
plt.figure(figsize = (20,20))
viz_1.loc[:,['On-Target','Goal Scored']].plot.bar()
plt.title('Relations between Goals and Shot On-Target on Every Team in FIFA World Cup')
plt.xlabel('Team')
plt.ylabel('Goal')
plt.show()
```

```python
# Relasi antara tendangan on-target dan tendangan masuk tidak saling bergantungan
# Dapat dilihat bagaimana pada negara German meskipun tendangan on-target tinggi
# Namun tendangan masuk hanya kurang dari 5
```

```python
plt.figure(figsize = (20,20))
sns.countplot(y=new_feature['Continent'], data=new_feature, order=new_feature['Continent'].value_counts().index)
plt.title('Total Countries in Every Continent FIFA World Cup')
plt.xlabel('Total Countries')
plt.ylabel('Continent')
plt.show()
```

```python
# Benua Eropa memiliki perwakilan negara paling banyak daripada benua lainnya. 
# Dengan didominasi benua Eropa maka peluang pemenang dalam pertandingan piala dunia
# Dimenangkan oleh benua Eropa
```

```python
viz_3 = pd.DataFrame(data_match.groupby(['Team']) ['Fouls Committed'].sum())
viz_3
```
